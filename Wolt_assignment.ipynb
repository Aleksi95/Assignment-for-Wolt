In this assignment we look at the data given, describe and try to visualize it. First we will look at the data.

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

dat = pd.read_json('example_2.json')
dat.info()
dat.timestamp = pd.to_datetime(dat.timestamp, infer_datetime_format = True)

dat.sort_values(by='timestamp')
print(len(dat))
print('vegetarian' in dat.venue_tags)

#function to select subset and frequency by food type
def bytag(data, tag):
    tags = []
    for i in data.index:
        tags.append(tag in data.venue_tags[i])
    return data[tags]

sub = bytag(dat, 'vegetarian')
sub1 = bytag(dat, 'meat')
print('frequency of orders containin tag vegetarian: ',len(sub))
print('frequency of orders containin tag meat: ',len(sub1))

```

```python

print(dat.timestamp[0].weekday_name == 'Monday')

#function to select by weekday
def byDay(data, day):
    d = []
    for i in data.index:
        d.append(data.timestamp[i].weekday_name == day)
    return data[d]

#subMon = byDay(dat, 'Monday')

days = ['Monday', 'Tuesday', "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
freqs = []

for day in days:
    freqs.append(len(byDay(dat, day)))

print(freqs)
plt.bar(days, freqs)
    
```

We can see that sunday is the busiest day for orders. Let's see how it changes by tags.

```python
freqs = []

for day in days:
    freqs.append(len(byDay(bytag(dat, 'hangover'), day)))

print(freqs)
plt.bar(days, freqs)
```

We can see that orders containing the tag 'hangover' peak at sunday and second is saturday.

Next let's look at weekly orders.

```python
week = []
for i in dat.index:
    week.append(dat.timestamp[i].week)

dat['week'] = pd.Series(week, index=dat.index)

orders = pd.DataFrame(dat.week.value_counts())
orders = orders.reset_index()
orders.columns = ['week', 'orders']

plt.scatter(orders['week'], orders['orders'])


```

For more descriptive data, let's look at daily orders, and construct a new dataset, based on the number of orders per each day.

```python
day = []
for i in dat.index:
    day.append(dat.timestamp[i].dayofyear)
    
dat['day'] = pd.Series(day, index=dat.index)
    
ordersPerDay = pd.DataFrame(dat.day.value_counts())
ordersPerDay = ordersPerDay.reset_index()
ordersPerDay.columns = ['day', 'orders']


plt.scatter(ordersPerDay['day'], ordersPerDay['orders'])
```

```python
avgDistance = []

for d in ordersPerDay['day']:
    mean = np.mean(dat[dat.day == d]['distance'])
    avgDistance.append(mean)

ordersPerDay['avgDistance'] = avgDistance

weekday = []

for d in ordersPerDay['day']:
    weekday.append(dat[dat.day == d]['timestamp'].iloc[0].weekday_name)
    

ordersPerDay['weekday'] = weekday


ordersPerDay
```

```python

#convert categorical weekday variable to numeric
for i in ordersPerDay.index:
    if ordersPerDay.weekday[i] == 'Monday':
        weekday[i] = 1
    if ordersPerDay.weekday[i] == 'Tuesday':
        weekday[i] = 2
    if ordersPerDay.weekday[i] == 'Wednesday':
        weekday[i] = 3
    if ordersPerDay.weekday[i] == 'Thursday':
        weekday[i] = 4
    if ordersPerDay.weekday[i] == 'Friday':
        weekday[i] = 5
    if ordersPerDay.weekday[i] == 'Saturday':
        weekday[i] = 6
    if ordersPerDay.weekday[i] == 'Sunday':
        weekday[i] = 7

        
ordersPerDay['weekday'] = weekday


ordersPerDay
        

```

Now let's try training a linear model to predict the number of daily orders. We will use the day, average distance and the weekday variables as explanatory variables.

```python
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error, r2_score

x = np.array(ordersPerDay[['day', 'avgDistance', 'weekday']], dtype='float')
y = np.array(ordersPerDay['orders'], dtype='float')

x = sm.add_constant(x)
np.random.seed(42)
testI = np.random.randint(0, high=len(x), size=len(x)//5)

xtrain = x[-testI,:]
ytrain = y[-testI]
xtest = x[testI,:]
ytest = y[testI]


model = sm.OLS(ytrain, xtrain).fit()
predictions = model.predict(xtest) # make the predictions by the model

print(predictions)

# Print out the statistics
model.summary()
```

The weekday seems to predict the number of orders best. Let's add the weather variable to our data.

```python
weather = []

for d in ordersPerDay['day']:
    weather.append(dat[dat.day == d]['weather'].iloc[0])
    
ordersPerDay['weather'] = weather

print(ordersPerDay.weather.value_counts())
```

```python
#convert to numeric
for i in ordersPerDay.index:
    if ordersPerDay.weather[i] == 'sunny':
        weather[i] = 1
    if ordersPerDay.weather[i] == 'mostly_sunny':
        weather[i] = 2
    if ordersPerDay.weather[i] == 'overcast':
        weather[i] = 3
    if ordersPerDay.weather[i] == 'rain':
        weather[i] = 4

ordersPerDay['weather'] = weather

ordersPerDay
```

Now let's fit the model again to the new data.

```python
x = np.array(ordersPerDay[['day', 'avgDistance', 'weekday', 'weather']], dtype='float')
y = np.array(ordersPerDay['orders'], dtype='float')

x = sm.add_constant(x)
np.random.seed(42)
testI = np.random.randint(0, high=len(x), size=len(x)//5)

xtrain = x[-testI,:]
ytrain = y[-testI]
xtest = x[testI,:]
ytest = y[testI]


model = sm.OLS(ytrain, xtrain).fit()
predictions = model.predict(xtest) # make the predictions by the model

print(predictions)

# Print out the statistics
model.summary()
```

Here we can see that the effect of the weather variable is not statistically significant. However, the data is from the summer months so most of the days were sunny. From looking at the data, we can see orders increasing when it is raining, so a better model could have taken this into account. In conclusion, based on the data in hand, the day of the week is best at predicting orders per day, and we can see orders clearly increasing by weekends, especially sundays.
